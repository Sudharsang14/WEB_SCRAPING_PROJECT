```python
# Import BeautifulSoup for parsing HTML
from bs4 import BeautifulSoup

# Import requests to send HTTP requests
import requests
```

```python
# URL of the website to scrape
url="https://www.ipl.com/cricket/news/ipl-2025-schedule-full-list-of-matches-fixtures-dates-venues-and-timings/"
# Send GET request to the website
page=requests.get(url)
```

```python
#BeautifulSoup
soup=BeautifulSoup(page.text,'html')
print(soup)
```

```python
# Find all tables in the webpage and select the first table
box=soup.find_all('table')[0]
box
```

```python
initial= box.find_all('th')
below=[o.text.strip() for o in initial]
below
```

```python
# Import pandas to store scraped data in DataFrame
import pandas as pd
# Create an empty DataFrame with column names
df1=pd.DataFrame(columns=below)
df1
```

```python
# Extract data from table body
for tbody in box.find_all('tbody'):        # Loop through each <tbody>
    for tr in tbody.find_all('tr'):        # Loop through each row <tr>
        cols=[td.text.strip() for td in tr.find_all('td')]
        length=len(df1)                    # Get current length of DataFrame (next row index)
        df1.loc[length]=cols               # Insert row into DataFrame
```

```python
df1.to_csv()                             #save dataframe to csv
df1
```
